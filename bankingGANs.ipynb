```python
from ctgan import CTGANSynthesizer
from ctgan import load_demo
import pandas as pd 
```

```python
data = pd.read_csv("C:/Users/mmrha/Desktop/cmu/Independent study/python/bankingmin.csv")

data.head()
```

```python
# Names of the columns that are discrete
discrete_columns = [
    "job",
    "marital",
    "education",
    "default",
    "housing",
    "loan",
    "contact",
    "month",
    "day_of_week",
    "campaign",
    "previous",
    "poutcome",
    "emp.var.rate",
    "cons.price.idx",
    "cons.conf.idx",
    "euribor3m",
    "nr.employed"
]

ctgan = CTGANSynthesizer(epochs=5000)
ctgan.fit(data, discrete_columns)

```

```python
# Synthetic copy
samples = ctgan.sample(40000)
```
